```python
Prifix = ["Mr.", "Mrs.", "Ms."]

Education = ["BCA", "BE", "B-Tech", "BSc", "BCS", "B-Com", "BBA", "BA",
             "MA", "MCA", "ME", "M-Tech", "MSc", "M-Com", "MBA"]

Tech_Skills = ["Python", "Java", "C", "C++", "C#", "JavaScript", "HTML", "CSS", "React",
               "Excel", "SQL", "PowerBI", "Tableau", "iOS Swift", "Android Studio",
               "AI", "ML", "Linux", "cloudcomputing", "cyber security"]

Soft_Skills = ["Problem Solving", "Communication Skills", "Teamwork Skills", "Critical Thinking",
               "Leadership", "Time Management", "Adaptability", "Attention To Details"]

Duration = ["4 Month", "6 Month", "8 Month", "12 Month"]


def choose_option(options, line):
    print(f"\n{line}")
    for i, opt in enumerate(options, start=1):
        print(f"{i}. {opt}")
    choice = int(input("Enter your choice number: "))
    return options[choice - 1]


def choose_multiple(options, line):
    print(f"\n{line} (Choose by entering comma-separated numbers)")
    for i, opt in enumerate(options, start=1):
        print(f"{i}. {opt}")
    choices = input("Enter your choice numbers (comma separated): ").split(",")
    result = []
    for c in choices:
        idx = int(c.strip()) - 1
        if 0 <= idx < len(options):
            result.append(options[idx])
    return result


def generate_resume():
    prifix = choose_option(Prifix, "Choose Your Prefix")
    name = input("Enter your Name: ")
    git = input("Enter your Github Profile URL: ")
    email = input("Enter Your Email: ")
    mobile_number = input("Enter Your Mobile Number: ")
    location = input("Enter Your Location: ")
    
    summary = input("Enter a brief professional summary: ")

    ssc_percentage = float(input("Enter Your SSC Percentage: "))
    hsc_percentage = float(input("Enter Your HSC Percentage: "))
    education = choose_option(Education, "Choose Your Highest Education:")
    passing_year = int(input(f"Enter {education} Passing Year: "))
    education_percentage = float(input(f"Enter {education} Percentage / CGPA: "))

    tech_skills = choose_multiple(Tech_Skills, "Choose Your Technical Skills:")
    soft_skills = choose_multiple(Soft_Skills, "Choose Your Soft Skills:")

    project = input("Have You Done Any Project (Enter Yes or No): ").strip().lower()
    if project == "yes":
        projects = input("Enter Your Project Title(s) (comma separated): ").split(",")
        projects = [p.strip() for p in projects if p.strip()]
    else:
        projects = []

    cert = input("Have You Done Any Certificate Course (Enter Yes or No): ").strip().lower()
    if cert == "yes":
        certificates = input("Enter Certificates You Have (comma separated): ").split(",")
        certificates = [c.strip() for c in certificates if c.strip()]
        duration = choose_multiple(Duration, "Choose Certification Duration(s):")
    else:
        certificates = []
        duration = []

    experience = input("Have You Any Experience (Enter Yes or No): ").strip().lower()
    if experience == "yes":
        experience_details = input("Enter Your Experience in Brief (comma separated): ").split(",")
        experience_details = [e.strip() for e in experience_details if e.strip()]
    else:
        experience_details = []


    print("\n" + "=" * 60)
    print(f"    {(prifix + ' ' + name).upper():^50}")
    print("=" * 60)
    print(f"⁌ Email: {email:<25}  ⁌ Mobile: {mobile_number}")
    print(f"⁌ Location: {location:<20}  ⁌ Github Profile: {git}")
    print("=" * 60)


    print("\n⁌ SUMMARY")
    print("-" * 60)
    words = summary.split()
    line = ""
    for word in words:
        if len(line) + len(word) + 1 <= 60:
            line += word + " "
        else:
            print(line.strip())
            line = word + " "
    if line:
        print(line.strip())

    if experience_details:
        print("\n⁌ EXPERIENCE")
        print("-" * 60)
        for ex in experience_details:
            print(f"• {ex}")

    if projects:
        print("\n⁌ PROJECTS")
        print("-" * 60)
        for pro in projects:
            print(f"• {pro}")
 



    
    print("\n⁌ EDUCATION")
    print("-" * 60)
    print(f"{education} ({passing_year}) - {education_percentage} CGPA")
    print(f"HSC: {hsc_percentage}%  | SSC: {ssc_percentage}%")

    if certificates:
        print("\n⁌ CERTIFICATIONS")
        print("-" * 60)
        for cert_item, dur in zip(certificates, duration):
            print(f"• {cert_item} – {dur}")

    print("\n⁌ TECHNICAL SKILLS")
    print("-" * 60)
    for skill in tech_skills:
        print(f"• {skill}")

    print("\n⁌ SOFT SKILLS")
    print("-" * 60)
    for skill in soft_skills:
        print(f"• {skill}")

   
    print("=" * 60)
    print("=" * 60)
    print("Resume Generated Successfully")


print("Welcome To Resume Generator")
while True:
    generate_resume()
    again = input("\nDo you want to generate another resume? (yes/no): ").strip().lower()
    if again != "yes":
        print("Thanks for using Resume Generator!")
        break
```
