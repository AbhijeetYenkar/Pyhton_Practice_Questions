```python
# fill the NaN values using the Median
df['historical_sales'] = df['historical_sales'].fillna(df['historical_sales'].median()) 
```

```python
# fill the NaN value with the forward fill
df['delivery_delay_days'] = df['delivery_delay_days'].ffill()
```

```python
# Replace Binalry format  into yes and no,
df['festival_season'] = df['festival_season'].replace({1:'yes',0:'No'})
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["stock_out_days"] = df["stock_out_days"].round().astype(int)
```

## 8. Are stock-outs more frequent during festival months?
