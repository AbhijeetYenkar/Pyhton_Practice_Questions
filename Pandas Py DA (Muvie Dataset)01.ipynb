```python
'''
26.fillna() - using mean
df['column_name'] = df['column_name'].fillna(df['column_name'].mean())
Fills missing values with the column mean
'''
df['mood_score'] = df['mood_score'].fillna(df['mood_score'].mean())
```

```python
'''
27.fillna() - using median
df['column_name'] = df['column_name'].fillna(df['column_name'].median())
Fills missing values with the column median
'''
df['meditation_sessions'] = df['meditation_sessions'].fillna(df['meditation_sessions'].median())
```

```python
'''
28.fillna() - using mode
df['column_name'] = df['column_name'].fillna(df['column_name'].mode()[0])
Fills missing values with the column median
mode() returns a Series. Use [0] to get the most frequent value.
'''
df['meditation_sessions'] = df['meditation_sessions'].fillna(df['meditation_sessions'].mode()[0])
```
