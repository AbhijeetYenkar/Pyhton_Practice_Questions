```python
def standard_daviation(a):
    # total number of elements
    n = 0
    for _ in a:
        n += 1

    # sum of numbers
    total = 0
    for value in a:
        total += value

    # calculate mean of numbers
    average = total / n

    # each data point, subtract the mean and square the result
    square_diff_sum = 0
    for value in a: 
        result = value - average
        square = result * result
        square_diff_sum += square

    # calculate variance 
    variance = square_diff_sum / n

    # take the square root to get standard deviation
    root = variance ** 0.5

    return root


data = [8, 5, 7, 9, 6]
r = standard_daviation(data)
print(r)
```

```python
def variance(x,y):
    # count the length of both list x and y
    n_x = 0
    for _ in x:
        n_x += 1
    n_y = 0
    for _ in y:
        n_y += 1

    # check if both list are same
    if n_x != n_y:
        return "both list are same"

    # calculate sum of element
    total_x = 0
    for value in x:
        total_x += value

    total_y = 0
    for value in y:
        total_y += value

    # calculate mean of list's
    mean_x = total_x / n_x
    mean_y = total_y / n_y

    # calculate variance using stddev
    sum_diif_square = 0
    i = 0
    while i < n_x:
        diff_x = x[i] - mean_x
        diff_y = y[i] - mean_y
        sum_diif_square += diff_x * diff_y
        i += 1

    variance = sum_diif_square / n_x
    return variance

x = [1,2,3,4,5]
y = [6,7,8,9,8]
result = variance(x,y)
print(result)
        
    
```

```python
def vector_3d(a1,b1,c1,a2,b2,c2):
    Dot = (a1 * a2) + (b1 * b2) + (c1 * c2)
    if Dot == 0:
        return "Perpendicular Vector"
    else:
        return "Not Perpendicular Vector"
vector_3d(2,3,4,5,6,7)
```

```python
def GCD(a,b):
    if b == 0:
        return a
    else:
        return GCD(b , a % b)
def LCM(a,b):
    if a == 0 or b == 0:
        return 0
    else:
        return (a * b) // GCD(a, b)


GCD(12,22)
```

```python
def GCD_n(a,b):
    if b == 0:
        return a
    else:
        return GCD_n(b , a % b)

GCD_n(30,60)
```

```python
def sum_n(n):
    if n == 0:
        return 0
    if n % 2 != 0:
        return n + sum_n(n - 1)
    else:
        return sum_n(n - 1)
  

sum_n(30)
```

```python
def product_of_digit(n):
    if n == 0:
        return 1
    else:
        return (n % 10) * product_of_digit(n // 10) 

product_of_digit(123)
```

```python
def reverce(n,r_n = 0):
    if n == 0:
        return r_n


    last_digit = n % 10
    return reverce(n // 10, r_n * 10 + last_digit) 

reverce(123)
```
